```python
import pandas as pd
import ta
import matplotlib.pyplot as plt
import numpy as np
```

```python
ibm_prices = pd.read_csv('https://query.data.world/s/ovn37kwnxqy4txkj4f7s662wxz5ppu')
ibm_prices_subset = ibm_prices[0:300]
del ibm_prices
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
ibm_prices_subset.close.describe()
```

```python
plt.plot(ibm_prices_subset.close)
```

```python
indicator_bb = ta.volatility.BollingerBands(close=ibm_prices_subset.close, n=20, ndev=3)
ibm_prices_subset['bb_bbm'] = indicator_bb.bollinger_mavg()
ibm_prices_subset['bb_bbh'] = indicator_bb.bollinger_hband()
ibm_prices_subset['bb_bbl'] = indicator_bb.bollinger_lband()
```

```python
ma = ibm_prices_subset['bb_bbm'].to_numpy()
upper_band = ibm_prices_subset['bb_bbh'].to_numpy()
lower_band = ibm_prices_subset['bb_bbl'].to_numpy()
x_axis = ibm_prices_subset.id.to_numpy()
price = ibm_prices_subset.close.to_numpy()
```

```python
plt.plot(x_axis, price)
plt.plot(x_axis, upper_band)
plt.plot(x_axis, lower_band)
plt.legend()
```
